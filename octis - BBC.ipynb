```python
import sys, numpy as np, pickle
if '/home/sagemaker-user/octis/OCTIS' not in sys.path: sys.path.append('/home/sagemaker-user/octis/OCTIS') 

from octis.models.LDA import LDA
from octis.models.LDA_gibbs import LDA_gibbs
from octis.models.LDA_tomopy import LDA_tomopy
from octis.models.NMF_scikit import NMF_scikit

from octis.dataset.dataset import Dataset
from octis.evaluation_metrics.diversity_metrics import TopicDiversity
from octis.evaluation_metrics.coherence_metrics import Coherence

from octis.optimization.optimizer import Optimizer
from skopt.space.space import Real, Integer, Categorical

from warnings import filterwarnings
filterwarnings('ignore')
```

```python
#loadData
bbc_dataset = Dataset()
bbc_dataset.fetch_dataset('BBC_News')

coherence = Coherence(texts=bbc_dataset.get_corpus(), topk=10, measure='c_v')
```

NMF_scikit

```python
nmf_model_20 = NMF_scikit(num_topics=20, init='nndsvdar', alpha=0.0)
nmf_output_20 = nmf_model_20.train_model(bbc_dataset)
print('Topics:',*[' '.join(x) for x in nmf_output_20['topics']],sep='\n')
print(f'\nOverall Coherence Score: {coherence.score(nmf_output_20)}')
```

```python
#preprocessData
import sys
if 'E:\\Rashi\\octis\\OCTIS' not in sys.path: sys.path.append('E:\\Rashi\\octis\\OCTIS') 

from octis.preprocessing.preprocessing import Preprocessing
import spacy, string
spacy.load('en_core_web_sm')
from spacy.lang.en import stop_words
stop_words=stop_words.STOP_WORDS

#datafiles
datafile = "E:\\Rashi\\datasets\\talk.religion.misc.txt\\talk.religion.misc.txt"
processed_datafile=source = "E:\\Rashi\\datasets\\talk.religion.misc.txt\\talk.religion.misc.processed.txt"
split_condition = 'Newsgroup: talk.religion.misc'
pickleFile_target="E:\\Rashi\\octis\\helperFiles\\talk.religion.misc.octisProcessedDataset.pkl"

#readData
with open(datafile,'r') as file:
    df=file.read()

#Pre-processing
documents=df.split(split_condition)[1:]

i=0
while i<len(documents):
    try:
        idx=documents[i].index('Subject')
        documents[i]=documents[i][idx:].replace('\n',' ').replace('\t',' ')
    except:
        print(f'breaks it. Subject missing in file {i}')
        break
    i+=1

documents_merged='\n'.join(documents)
with open(processed_datafile,'w') as file:
    file.write(documents_merged)

#octisPreprocessing
stop_words=list(stop_words)+['share','article','futurism','d', 'll', 'm', 'n', 's', 't', 've']
preprocessor = Preprocessing(vocabulary=None, max_features=None,
                             remove_punctuation=True, punctuation=string.punctuation,
                             lemmatize=True, stopword_list=stop_words,
                             min_chars=1, min_words_docs=50, encoding='latin-1')

dataset = preprocessor.preprocess_dataset(documents_path=processed_datafile) 

with open(pickleFile_target,'wb') as file:
    pickle.dump(dataset,file)
```

```python
with open(pickleFile_target,'rb') as file:
    dataset=pickle.load(file)

coherence = Coherence(texts=dataset.get_corpus(), topk=10, measure='c_v')
```
